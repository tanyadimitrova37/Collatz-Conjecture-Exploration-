## Explanation of the Collatz Conjecture and its Formulation.

The Collatz conjecture, also known as $3n + 1$  problem, is one of the most intriguing, yet unsolved problems of mathematics. The conjecture has perplexed scholars and enthusiasts alike since it was first proposed by Lothar Collatz in 1937.  It can be stated quite simply. Take any positive integer $n$. If $n$ is an even number, divide it by 2. If $n$ is odd, multiply it by 3 and add 1. Repeat the same operation with the result you get. The conjecture states that any sequence seems to get stuck in the 4, 2, 1 infinite loop, regardless of the initial number you chose. Stated formally:

   Let $T: \mathbb{Z} \rightarrow \mathbb{Z}$ be defined by:
$$T(n) = 
\begin{cases} 
\frac{n}{2} & \text{if } n \text{ is even} \\
3n+1 & \text{if } n \text{ is odd}
\end{cases}$$

The conjecture postulates that, given $n \in \mathbb{N}$, the trajectory: $n, \; T(n), \; T^2(n),  \ldots,$ eventually reaches the cycle 4, 2, 1.

## Review of previous attempts to prove or disprove the conjecture, highlighting key insights

A fundamental question about the Collatz sequence is whether it always results in a finite loop, specifically the 4-2-1 cycle, regardless of the starting number. The conjecture states that it does, but this has never been proven for all natural numbers. 

As of 2020, the conjecture has been checked by computer for all starting values up to $2^{68} ≈ 2.95×10^{20}$. All values tested so far converge to 1 (Barina, 2020).

### Terras

One notable contribution is by Terras (1976), who initiates his analysis by defining the function $F(k)$, which measures the probability that a randomly selected integer from 1 to $m$ has a stopping time greater than $k$ as $m$ approaches infinity. The function is formally defined as:
$$F(k) = \lim_{m \to \infty} \left( \frac{1}{m} \#\{n \leq m|\chi(n) \geq k\} \right)$$
$$\lim_{k \to \infty}F(k) = 0$$

The implication of $F(k) = \lim_{m \to \infty}$ would be that as k increases, the proportion of numbers satisfying $\chi(n) \geq k$ pproaches zero, suggesting that most sequences eventually fall below any given $k$ after sufficiently large numbers, supporting the conjecture that all numbers eventually lead to the repeating cycle 4, 2, 1. This implies that the likelihood of an infinite stopping time in the Collatz sequence is virtually nonexistent. Terras's innovative approach not only enriches the theoretical landscape but also sets a precedent for using statistical methods in the exploration of other complex mathematical problems.

### Eliahou

Eliahou's key finding in 1993 establishes a minimum possible cycle length of 17,087,915 for values greater than $2^{40}$. This number is significantly high, suggesting that any cycle, if one exists other than the known trivial cycle (1, 2, 4), must be very long, implying that finding such a cycle computationally has very low, if not 0, probability. This strongly indicates that a counterexample to the Collatz conjecture likely does not exist.

### Tao

Tao's 2019 work on the Collatz conjecture represents the most significant progress in recent times toward understanding this problem. In his paper, Tao establishes that for any function $f(x)$ that approaches positive infinity as $x$ approaches infinity, the smallest value in the Collatz orbit of $x$ will be below $f(x)$ for nearly all $x$. In other words, almost all numbers would end up smaller than any arbitrary function $f(x)$ so long as that function goes to infinity as $x$ goes to infinity. What this means is that for **almost** all numbers you can guarantee that there's an arbitrarily small number somewhere in its sequence. That small number, as a seed, will get smaller, and so on and so forth, until it reaches 1.

## Introduction to Collatz sequences and their recursive properties

Let us examine how a simple Collatz sequence behaves, by applying the rules outlined above. For example, take the starting number **$n_0$ = 6**. It is even so we divide it by 2, which gives **$n_1$ = 3** as a result. 3 is an odd number so we multiply it by 3 and add 1. This leads us to **$n_2$ = 10** as the result. We divide that by 2 and we get **$n_3$ = 5**. Following those rules we get then get the following numbers: **16, 8, 4, 2, 1 ($n_4, n_5, n_6, n_7$)**. However, 1 is an odd number. So if we keep applying the rules, we go back to **4**, which leads us back to **2** and then **1**, **over and over again**. In this particular example, it took us 9 iterations to get to 1. The number of iterations that is taking the sequence to reach 1 is also called **stopping time** (Lagarias, Jeffrey C., 1985). However, it seems that no matter which number we chose as a staring point, given that it is a positive integer, belonging to the set of natural numbers, we always reach the 4, 2, 1 loop. No one yet has been able to prove the conjecture, though. Or disprove it, for that matter.

The Collatz sequences are a classic example of recursive process in the sense that each term $T(n)$ in the sequence depends directly on the properties of the preceding term $n$. Moreover, the created sequences are highly dependent on the initial integer and are characterized by unpredictable lengths and behaviors, showing sensitivity similar to that seen in chaotic systems. Therefore, the sequences can grow rapidly through runs of odd numbers and reduce just as quickly when even numbers appear, highlighting the recursive and dynamic nature of the process. 

For example, if you take $n_0$ = 26 as the starting integer of the sequence, it will take 10 steps to reach 1 (*Fig. 1*) and the highest number from the sequence is 40. However, if you take 27, a total of 111 iteration will be done for the sequence to reach 1 (*Fig. 2*) and the highest number reached is as high as 9232: 

$$27, 82, 41, 124, 62, 31, 94, 47, 142, 71, 214, 107, 322, 161, 484, 242, 121, 364, 182, 91, 274, 137, 412, 206, 103, 310, 155, 466, 233, 700, 350, 175, 526, 263, 790, 395, 1186, 593, 1780, 890, 445, 1336, 668, 334, 167, 502, 251, 754, 377, 1132, 566, 283, 850, 425, 1276, 638, 319, 958, 479, 1438, 719, 2158, 1079, 3238, 1619, 4858, 2429, 7288, 3644, 1822, 911, 2734, 1367, 4102, 2051, 6154, 3077, 9232, 4616, 2308, 1154, 577, 1732, 866, 433, 1300, 650, 325, 976, 488, 244, 122, 61, 184, 92, 46, 23, 70, 35, 106, 53, 160, 80, 40, 20, 10, 5, 16, 8, 4, 2, 1$$

This demonstrates clearly that the length of the path to the cycle and the highest number reached can vary significantly depending on the starting number.

```python
def collatz_seq(n):
    sequence = []
    while n != 1:
        sequence.append(n)
        if n % 2 == 0:
            n //= 2
        else:
            n = 3 * n + 1
    sequence.append(1)
    return sequence

def plot_collatz_comparisons(start_nums):
    plt.figure(figsize = (14, 6))

    for index, num in enumerate(start_nums):
        seq = collatz_seq(num)
        steps = len(seq)
        highest_num = max(seq)
        
        plt.subplot(1, 2, index + 1)
        plt.plot(seq, marker = 'o')
        plt.title(f'(Fig. {index + 1}) Collatz Sequence Starting from {num} (Stopping time = {steps})\n Highest Number Reached: {highest_num}', linespacing = 2, pad = 20)
        plt.xlabel('Step Number')
        plt.ylabel('Sequence Value')
        plt.grid(True)
    
    plt.tight_layout()
    plt.show()

# Comparing the sequences for 26 and 27
plot_collatz_comparisons([26, 27])
```

# Computational Exploration and Visualization

## Investigation of the convergence properties of Collatz sequences.

Over the years, the conjecture has been tested on a vast range of numbers using computer algorithms. To date, every tested integer has eventually led to the cycle 4, 2, 1, providing a substantial body of empirical evidence supporting the conjecture. These computational experiments have covered extremely large numbers, far beyond typical human checking, and all have converged as predicted.

The convergence to the cycle 1, 2, 4 can be intriguingly consistent. Once a sequence reaches the number 1, it follows the path 1 → 4 → 2 → 1, and this cycle repeats indefinitely. This specific cycle's inevitability in tested cases lends credence to the conjecture's validity.

We can test this ourselves, even though we have some technical limitations. By running the function below, we can see that for any random integer between 1 and 1 000 000, we reach the cycle 4, 2, 1, and we go through it at least 1 000 000 times. You can test this with a different start number range and cycle limit.

```python
def collatz_cycle_counter(start):
    n = start
    cycle_count = 0
    sequence = []
    
    while cycle_count < 1000000:
        sequence.append(n)
        
        if n % 2 == 0:
            n = n // 2
        else:
            n = 3 * n + 1
                
        if sequence[-3:] == [4, 2, 1]:
            cycle_count += 1
            # Resetting the sequence to start counting a new cycle. From now on, the process will be repetitive: sequence = [4, 2, 1].
            sequence = []
    
    # Once we reach 1000000 cycles
    return print(f"Reached {cycle_count} cycles of 4, 2, 1 starting from {start}")

start = random.randint(1, 1000000)
collatz_cycle_counter(start)
```

The ultimate down trend in Collatz sequences is interesting. Considering there are infinite amount of numbers, half of them should be odd and the other half should be even. If that is the case, it would be intuitive to say that the sequences should grow instead of shrink, given that each odd number is multiplied by 3 and even numbers are only divided by 2.

However, each odd number which we triple and add 1 to it, will result in an even number. Therefore, the next step would be to halve it. Here is the "shortcut" of the Collatz function (Lohia, 2022):

$$T(n) = 
\begin{cases} 
\frac{3n + 1}{2}, & \text{if } n \equiv 1 \pmod{2} \\
\frac{n}{2}, & \text{if } n \equiv 0 \pmod{2}
\end{cases} 
$$

We can deduce that odd numbers are not actually tripled by $3x + 1$ but are increased by a factor of about $\frac{3}{2}$. We can more or less neglect the $+1$ as it would be insignificant for large numbers.

To go deeper in that, all odd numbers, will result in an even number, after applying the $3x+1$ operation. From there, $\frac{1}{2}$ of the times, we will get to an odd number on iteration number 2. $\frac{1}{4}$ of the time, we would get to an odd number by dividing by 4. Consequently, an $\frac{1}{8}$ of the time, we can divide by 8 to get to an odd number. A $\frac{1}{16}$ on the time we would have to divide by 16 to get to an odd number. And so on and so forth.

$$
\left(\frac{3}{2}\right)^{0.5} \left(\frac{3}{4}\right)^{0.25} \left(\frac{3}{8}\right)^{0.125} \left(\frac{3}{16}\right)^{0.0625} \ldots
$$

Represented mathematically:

$$\prod_{i=1}^{\infty} \left(\frac{3}{2^i}\right)^{\frac{1}{2^i}} = \prod_{i=1}^{\infty} \left(3\right)^{\frac{1}{2^i}} \cdot \prod_{i=1}^{\infty} \left(\frac{1}{2^i}\right)^{\frac{1}{2^i}}$$


$$\prod_{i=1}^{\infty} \left(3\right)^{\frac{1}{2^i}} = 3$$

$$\prod_{i=1}^{\infty} \left(\frac{1}{2^i}\right)^{\frac{1}{2^i}} = \frac{1}{4}$$

$$\prod_{i=1}^{\infty} \left(\frac{3}{2^i}\right)^{\frac{1}{2^i}} = \frac{3}{4}$$


It turns out that, to get to one odd number to the next one, we multiply, on average, by $\frac{3}{4}$ (Lohia, 2022). This is less than one, meaning that it would make sense for the sequence to shrink over time, rather than grow.

### Odd to Even Ratio

We can explore further by examining the odd to even ratio in the Collatz sequences. By plotting the distribution of the ratios (Fig. 6), we can see that the histogram shows a left-skewed distribution, in which the tail highlights sequences where even numbers significantly outnumber odd numbers. A peak near 0.6 indicates that many sequences have a ratio of odd numbers to even numbers around 0.6. This suggests that, generally, the number of odd numbers in these sequences is less than the number of even numbers.

The mean ratio of 0.4828 suggests that, on average, there are slightly more than two even numbers for every odd number across all sequences examined, which is consistent with the distribution analysis.

Exploring further (Fig. 7), a correlation coefficient of 0.02 is very close to zero, indicating that there is virtually no linear relationship between the starting numbers and their corresponding odd-to-even ratios. This suggests that knowing the starting number gives almost no information about what the odd-to-even ratio will be. There is one apparent outlier which is actually the number 2, for which the ratio is actually 1.0. If we remove that, the correlation is still around 0.02.

```python
def count_odd_even_numbers(limit_number):
    data = []
    for i in range(2, limit_number + 1): #Skipping 1 because of division by 0
        sequence = collatz_seq(i)
        odd_count = sum(1 for x in sequence if x % 2 != 0)
        even_count = sum(1 for x in sequence if x % 2 == 0)
        
        ratio = odd_count / even_count
        data.append([i, odd_count, even_count, ratio])
    return data

# Generating data and putting it into a dataframe
limit_number = 100000
data = count_odd_even_numbers(limit_number)

columns = ['Starting Number', 'Odd Count', 'Even Count', 'Odd to Even Ratio']
df = pd.DataFrame(data, columns=columns)

df.to_csv('collatz_odd_even_ratio.csv')
```

```python
ratios = df['Odd to Even Ratio']
mean_ratio = ratios.mean()

plt.figure(figsize = (10, 6))
plt.hist(ratios, bins = 50, alpha = 0.7, color = 'red', label = "Distribution")
plt.title('(Fig. 6) Distribution of Odd to Even Ratios', pad = 20)
plt.xlabel('Odd to Even Ratio')
plt.ylabel('Frequency')
plt.axvline(mean_ratio, color ='blue', linestyle = '--', linewidth = 1, label = "Mean")
plt.legend()
plt.show()

print(f"The mean ratio is: {mean_ratio}")
```

```python
correlation_coef = df['Starting Number'].corr(df['Odd to Even Ratio'])

plt.figure(figsize=(10, 6))
plt.scatter(df['Starting Number'], df['Odd to Even Ratio'], alpha=0.5)
plt.title('(Fig. 7) Correlation between Starting Number and Odd to Even Ratio', pad = 20)
plt.xlabel('Starting Number')
plt.ylabel('Odd to Even Ratio')
plt.show()

print(f"The correlation coefficient is: {correlation_coef}")
```

### Benford's Law

An alternative approach to examining the conjecture involves utilizing Benford's Law. This law, also referred to as the First-Digit Law, describes the frequency distribution of leading digits across various real-world numerical datasets, like population of countries (Wikimedia Commons, n.d.), market capitalization (Zgela, 2012), and the Fibonacci numbers (M. et al., 2017). According to Benford's Law, in a wide range of naturally occurring number collections, the first digit tends to be smaller (Berger & Hill, 2011). For example, in sets that obey the law, the number 1 appears as the leading digit about 30% of the time, much more often than would be expected if all digits were equally likely. The relationship between Collatz sequences and Benford's Law has been studied most by Kontorovich and Miller (2005) and and Lagarias and Soundararajan (2006).

If the numbers generated in Collatz sequences follow Benford's Law, it could imply a form of statistical regularity and predictability in the behavior of these sequences. This regularity might provide insights into the underlying dynamics of the numbers as they undergo the transformations specified by the Collatz function.

Studies on the topic by the aforementioned scientist are overall consistend in finding the Collatz sequences satisfy the Benford's Law conditions. The code below is designed to generate and examine the sequences for numbers ranging from 1 to 100,000, tallying how frequently the digits 1 through 9 appear as the first digit of each number within these sequences. The observation confirms that, consistent with Benford's Law, the digit 1 appears as the leading digit in approximately 30% of the cases (Fig. 8). The likelihood of each subsequent digit appearing as the leading digit decreases progressively, with the digit 9 occurring as the first digit in fewer than 5% of the cases.

```python
def get_leading_digit(num): # returning the first digit of a number
    return int(str(num)[0])

def lead_digit_count_dict(start, end): # returns a dictionary containing the occurences od each number from 1 to 9 as a leading digit (format--> lead digit : number of occurences)
    lead_digit_count = {}
    for i in range(start, end + 1):
        sequence = collatz_seq(i)
        for number in sequence:
            lead_digit = get_leading_digit(number)
            if lead_digit not in lead_digit_count:
                lead_digit_count[lead_digit] = 0
            lead_digit_count[lead_digit] += 1
    return lead_digit_count

def plot_benford_law(lead_digit_count): # it is also not necessary to put this in a function
    digits = sorted(lead_digit_count.keys())
    counts = [lead_digit_count[digit] for digit in digits]
    total_counts = sum(counts)
    frequences = [count / total_counts for count in counts]
    
    plt.figure(figsize = (10, 6))
    plt.bar(digits, frequences, tick_label = digits, color = 'skyblue')
    plt.xlabel('Lead Digit')
    plt.ylabel('Frequency')
    plt.title(f'(Fig. 8) Distribution of Lead Digits in Collatz Sequences from 1 to 100000', pad = 20)
    plt.show()
    
lead_digit_count = lead_digit_count_dict(1, 100000)
plot_benford_law(lead_digit_count)
```

While this law provides an interesting lens through which to examine the Collatz sequences, its application does not directly contribute to proving or disproving the conjecture. Instead, it offers a statistical perspective that might enrich understanding of the sequences' behavior or provide indirect insights into the nature of the numbers involved in the conjecture. The relationship between Benford's Law and the Collatz Conjecture remains largely an area for statistical exploration and mathematical curiosity rather than a definitive pathway to resolution of the conjecture

### Upper Bounds

The plot below shows the number of steps each numbers goes through before reaching the 4, 2, 1 cycle (Fig. 9). Even though we cannot identify a specific pattern here, we can still deduce that for the numbers up to 100 000, the stopping time rarely exceeds 350. For example, there are numbers close to 2000 that have a similar stopping time with numbers close to 10 000. This does not show a clear pattern or a relationship between the choice of a starting numbers and the total stopping time.

This observation does not necessarily imply that stopping times won't increase for starting numbers greater than 100,000. In fact, when we extend our examination to starting numbers up to 1,000,000, the highest stopping times observed are around 524. If we further expand the limit to 10,000,000, the upper range of stopping times increases to approximately 685. This trend suggests that as we test with progressively higher starting numbers, the maximum stopping times generally tend to increase. However, this trend does not establish a definitive relationship between the starting number and the length of the stopping time, indicating that while larger numbers may take longer to reach the cycle, the correlation is not strictly linear or predictable

```python
def collatz_steps(n):
    sequence = collatz_seq(n)
    steps = len(sequence)

    return steps

limit = 100000
numbers = np.arange(1, limit + 1)
stopping_times = np.array([collatz_steps(n) for n in numbers])

plt.figure(figsize = (10, 6))
plt.scatter(numbers, stopping_times, s = 1, color = 'green')
plt.title(f'(Fig. 9) Collatz Sequences Stopping Times for Numbers up to {limit}', pad = 20)
plt.xlabel('Number')
plt.ylabel('Stopping Time')
plt.grid(True)
plt.tight_layout()
plt.show()
```

We can observe the highest number reached for various sequences (Fig. 10), and it's evident that there isn't a straightforward relationship between the size of the starting number and the peak it reaches in its sequence. While it appears that larger starting numbers can potentially reach higher peaks, this trend is not consistent or predictable. There is no definitive correlation that reliably predicts the highest number based solely on the starting number.

```python
def collatz_sequence_max_num(n):
    sequence = collatz_seq(n)
    max_number = max(sequence)
    return max_number

limit = 10000
numbers = np.arange(1, limit + 1)
max_numbers = np.array([collatz_sequence_max_num(n) for n in numbers])

plt.figure(figsize = (7, 10))
plt.scatter(numbers, max_numbers, s = 1, color = 'green')
plt.title(f'(Fig. 10) Collatz Sequences Highest Number Reached for Seeds up to {limit}', pad = 20)
plt.xlabel('Number')
plt.ylabel('Max Value')
plt.ylim(0, 100000)
plt.grid(True)
plt.tight_layout()
plt.show()
```
